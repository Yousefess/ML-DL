```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_normal = stats.norm.rvs(
    size=100000,
    loc=0,
    scale=1
)
```

```python
sns.displot(
    data_normal,
    bins=100,
    kde=False
)
```

```python
data_uniform = stats.uniform.rvs(
    size=100000,
    loc=0,
    scale=10
)
```

```python
sns.displot(
    data_uniform,
    bins=100,
    kde=False
)
```

```python
data_bernoulli = stats.bernoulli.rvs(
    size=1000,
    p=0.2   # Win Probability
)
print(data_bernoulli)
```

```python
sns.displot(
    data_bernoulli,
    kde=False
)
```

```python
data_binom = stats.binom.rvs(
    size=1000,
    n=100,  # in each do 100 times / number in output is sum of 100 times
    p=0.2
)
print(data_binom)
```

```python
sns.displot(
    data_binom,
    bins=50,
    kde=False
)
```

```python
import numpy as np
```

```python
p = 1 / 6
data_multinomial = np.random.multinomial(
    n=6,
    pvals=[p, p, p, p, p, p],
    size=1000
)
print(data_multinomial)
```

```python
icu = 0.3
surgery = 0.2
el = 0.5
data_multinomial = np.random.multinomial(
    n=1000,
    pvals=[icu, surgery, el],
    size=365
)
print(data_multinomial)
```

```python
data_poisson = np.random.poisson(
    lam=7,
    size=1000
)
print(data_poisson)
```

```python
sns.displot(
    data_poisson,
    kde=False
)
```

```python
data_exponential = np.random.exponential(
    scale=2,
    size=1000
)
print(data_exponential)
```

```python
sns.displot(
    data_exponential,
    kde=False
)
```
